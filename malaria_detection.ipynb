#### CNN Model 2 - Technical details
The final proposed solution is a CNN model optimized with several techniques to enhance performance:

- **LeakyReLU Activation Function:** This activation function allows a small, non-zero gradient when the unit is not active, which helps mitigate the vanishing gradient problem.
- **Batch Normalization:** Normalizes the activations of each layer, maintaining a stable distribution of activations throughout the network and resulting in smoother gradients during backpropagation.
- **Dropout Layers:** Introduced after convolutional and fully connected layers to prevent overfitting by randomly setting a fraction of input units to zero during training.

The training process was enhanced using the following callbacks:
- **EarlyStopping:** Monitors validation loss and stops training if no improvement is observed after 5 epochs.
- **ModelCheckpoint:** Saves the best model based on validation accuracy.
- **ReduceLROnPlateau:** Reduces the learning rate by a factor of 0.5 if the validation loss does not improve after 3 epochs.
- **LearningRateScheduler:** Keeps the learning rate constant for the first 10 epochs and then exponentially decays it.

```python
# Storing the training path in a variable named DATADIR, and storing the unique categories/labels in a list

# Path of training data after unzipping
DATADIR_TEST = "/content/cell_images/test"
# Path of test data after unzipping
DATADIR_TRAIN = "/content/cell_images/train"
# Storing all the categories in 'CATEGORIES' variable
CATEGORIES = ["parasitized", "uninfected"]
```

```python
# function for extracting data in batches
def extract_data(datadir, batch_size=1):
    """
    Function for extracting data in batches
    need to load in batches to save on RAM.
    """
    extracted_data = []
    for category in CATEGORIES:
        path = os.path.join(datadir, category)
        class_num = category
        # Get the list of files in the directory
        files = os.listdir(path)
        num_files = len(files)
        # Calculate the number of batches
        num_batches = (num_files + batch_size - 1) // batch_size
        # Iterate over batches
        for i in range(num_batches):
            # Calculate start and end indices for the current batch
            start_idx = i * batch_size
            end_idx = min((i + 1) * batch_size, num_files)
            # Iterate over files in the current batch
            for file_name in files[start_idx:end_idx]:
                img_array = cv2.imread(os.path.join(path, file_name))
                extracted_data.append([img_array, class_num])
    return extracted_data
```

### <b>Check the minimum and maximum range of pixel values for train and test images

Let's define a function to calculate the minimum and maximum range of image sizes

```python
# function to get minimum and maximum image size considering both width and height
def get_min_max(dataset):
    min_width = min_height = np.inf
    max_width = max_height = -np.inf

    for image in dataset:
        num_pixels_h, num_pixels_w, _ = image.shape

        if num_pixels_w < min_width:
            min_width = num_pixels_w
        if num_pixels_w > max_width:
            max_width = num_pixels_w
        if num_pixels_h < min_height:
            min_height = num_pixels_h
        if num_pixels_h > max_height:
            max_height = num_pixels_h

    width_range = max_width - min_width
    height_range = max_height - min_height

    return min_width, max_width, min_height, max_height,width_range, height_range
```

###<b>Normalize the images

**Noramalization**
* Normalization makes the training faster and reduces the chances of getting stuck at local optima.
* In CNNs normalization helps to avoid exploding gradient problems.
* As we know image pixel values range from 0-255, here we are simply dividing all the pixel values by 255 to standardize all the images to have values between 0-1.

**Resize image**
* As we saw earlier the images size varies both in the test and training datatset
* Let's resize the images to 150x150 as it offers a good balance between computational efficiency and retaining enough details for CNN to learn the features.
* Let's keep the image aspect ratio and add padding. We could also perform a Resize with distortion, but let's keep it for later on for Model 3 with augmentation.
* Lets process the images in batches to save on RAM

```python
def resize_and_normalize_batch(images, normalize=True, IMG_SIZE = 150):
    """
    Resize a batch of images while maintaining the aspect ratio, add padding if necessary
    to match the target size, and optionally normalize the pixel values.

    Parameters:
        images: Batch of input image arrays with shape (num_samples, height, width, channels).
        normalize: Boolean indicating whether to normalize the pixel values. Default is True.
        IMG_SIZE: Target size for the resized and padded images.

    Returns:
        Resized, padded, and optionally normalized batch of images.
    """

    resized_and_padded_images = []
    for image in images:
        # Compute scaling factor for resizing
        scaling_factor = min(IMG_SIZE / image.shape[1], IMG_SIZE / image.shape[0])

        # Resize the image while maintaining aspect ratio
        resized = cv2.resize(image, None, fx=scaling_factor, fy=scaling_factor)

        # Compute padding
        pad_width = IMG_SIZE - resized.shape[1]
        pad_height = IMG_SIZE - resized.shape[0]

        # Apply padding to center the resized image
        top = (pad_height) // 2
        bottom = pad_height - top
        left = (pad_width) // 2
        right = pad_width - left

        padded = cv2.copyMakeBorder(resized, top, bottom, left, right, cv2.BORDER_CONSTANT)

        if normalize:
            # Normalize pixel values to range [0, 1]
            # Convert to float32 to ensure numerical stability during computations
            normalized = padded.astype(np.float32) / 255.0
            resized_and_padded_images.append(normalized)
        else:
            resized_and_padded_images.append(padded)

    return np.array(resized_and_padded_images)
```

```python
def resize_in_batches(images, batch_size, normalize=True, IMG_SIZE = 150):
    """
    Process images in batches: resize and pad each batch of images.

    Parameters:
        images
        batch_size
    Returns:
        resized and padded images processed in batches.
    """
    num_batches = len(images) // batch_size
    resized_batches = []

    for i in range(num_batches):
        batch = images[i * batch_size:(i + 1) * batch_size]
        resized_and_padded_batch = resize_and_normalize_batch(batch,normalize,IMG_SIZE)
        resized_batches.append(resized_and_padded_batch)

    # Process remaining images if they don't fit into a full batch
    if len(images) % batch_size != 0:
        batch = images[num_batches * batch_size:]
        resized_and_padded_batch = resize_and_normalize_batch(batch,normalize,IMG_SIZE)
        resized_batches.append(resized_and_padded_batch)

    return np.concatenate(resized_batches, axis=0)
```

### <b>Converting RGB to HSV of Images using OpenCV

Converting RGB images to HSV (Hue, Saturation, Value):
* Hue : type of color, such as RGB
* Saturation : intensity of the color
* Value : This represents the brightness of the color

It can be beneficial for various image processing tasks
* Simpler color-based segmentation: Helps to separate objects of different colors in a picture
* Invariance to illumination changes: Avoid the case when the lighting in a picture changes, making colors look different, focuse on actual color of the image.


```python
# import lib for handling conversion
import cv2
```

```python
def convert_rgb_to_hsv_batch(dataset, batch_size):
    """
    Convert RGB images in a batch dataset to HSV.

    Parameters:
        dataset 4D vector containing RGB images with shape (num_samples, height, width, 3)
        batch_size - size of each batch.

    Returns:
        batch dataset with images converted to HSV format
    """
    num_samples = len(dataset)
    num_batches = num_samples // batch_size
    remaining_samples = num_samples % batch_size

    hsv_batch = []

    # Process full batches
    for i in range(num_batches):
        batch = dataset[i * batch_size: (i + 1) * batch_size]
        batch_hsv = [cv2.cvtColor(img, cv2.COLOR_RGB2HSV) for img in batch]
        hsv_batch.extend(batch_hsv)

    # Process remaining images
    if remaining_samples > 0:
        remaining_batch = dataset[num_batches * batch_size:]
        remaining_hsv_batch = [cv2.cvtColor(img, cv2.COLOR_RGB2HSV) for img in remaining_batch]
        hsv_batch.extend(remaining_hsv_batch)

    return np.array(hsv_batch)
```

###<b> Processing Images using Gaussian Blurring

* Gaussian smoothing, is a process that involves averaging pixel values with their neighbors, weighted by their distance from the central pixel, using a Gaussian distribution.
* Benefits:
  * Noise Reduction - noise can be smoothed out by averaging pixels with their neighbors.
  * Enhance image structure, Remove small details - remove fine details or small textures from an image, focusing on larger structures. Detect the most important features.

**Note:**
* let's define a blur function
* hyperparameters:
  * kernel_size - size of the Gaussian kernel. It should be an odd number to ensure that there is a central pixel. The larger the value the smoother the blur transition, but more computational cost (common value 5,5).
  * sigma - controls the amount of blurring. A higher value results in more blurring. Let's select sigma = 0.0 as small sigma values will help with noise reduction while keeping the important details.

```python
def apply_gaussian_blur_batch(images, batch_size):
    """
    Apply Gaussian blur to a batch of images

    Parameters:
        images
        batch_size

    Returns:
        batch of blurred images.
    """
    kernel_size=(5, 5)
    sigma=0.0

    num_samples = len(images)
    num_batches = num_samples // batch_size
    remaining_samples = num_samples % batch_size

    blurred_images = []

    # Process full batches
    for i in range(num_batches):
        batch = images[i * batch_size: (i + 1) * batch_size]
        batch_blurred = [cv2.GaussianBlur(img, kernel_size, sigma) for img in batch]
        blurred_images.extend(batch_blurred)

    # Process remaining images
    if remaining_samples > 0:
        remaining_batch = images[num_batches * batch_size:]
        remaining_blurred_batch = [cv2.GaussianBlur(img, kernel_size, sigma) for img in remaining_batch]
        blurred_images.extend(remaining_blurred_batch)

    return np.array(blurred_images)
```

###<b>Bonus (Ensemble)</b>

Let's build an ensemble model based on the 3 best-performing models cnn_2, cnn_4, cnn_5, and see if it gives better results

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generating predictions from each model
predictions = []

for model in [cnn_2, cnn_4, cnn_5]:
    if model == cnn_4:
        predictions.append(model.predict(X_test_32))
    else:
        predictions.append(model.predict(X_test))

# Converting predictions to class labels
predictions = [np.argmax(pred, axis=1) for pred in predictions]

```

```python
# Combining predictions using max voting
ensemble_predictions = []

for i in range(len(X_test)):
    sample_predictions = [pred[i] for pred in predictions]
    counts = np.bincount(sample_predictions)
    majority_vote = np.argmax(counts)
    ensemble_predictions.append(majority_vote)

# Ensure ensemble_predictions is a numpy array
ensemble_predictions = np.array(ensemble_predictions)
```

```python
# Evaluate ensemble performance
ensemble_accuracy = accuracy_score(y_test_encoded.argmax(axis=1), ensemble_predictions)
ensemble_precision = precision_score(y_test_encoded.argmax(axis=1), ensemble_predictions)
ensemble_recall = recall_score(y_test_encoded.argmax(axis=1), ensemble_predictions)
ensemble_f1 = f1_score(y_test_encoded.argmax(axis=1), ensemble_predictions)

# Generate confusion matrix
conf_matrix = confusion_matrix(y_test_encoded.argmax(axis=1), ensemble_predictions)

# Print evaluation metrics
print("Ensemble Accuracy:", ensemble_accuracy)
print("Ensemble Precision:", ensemble_precision)
print("Ensemble Recall:", ensemble_recall)
print("Ensemble F1-score:", ensemble_f1)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", xticklabels=["parasitized", "uninfected"], yticklabels=["parasitized", "uninfected"])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```
